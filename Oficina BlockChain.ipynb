## 3 - BlockChain e o problema que ele resolve

Como o nome indica BlockChain é uma corrente de blocos que contêm informações. Essa técnica foi descrita pela primeira vez em 1991 por um grupo de pesquisadores que tinham a intenção original de carimbar um documento digital dificultando a alteração das informações e de que o dados fossem misturados em outros. Exemplo Tabelião. Mas Permaneceu desconhecida atá ser adaptado por Satoshi Nakamoto em 2009 para criar a criptompeda digital Bitcoin.

Um blockchain é uma ledger distribuida aberta a qualquer um e que possui a característica de que uma vez um dado seja gravado dentro do BlockChain torna-se muito difícil de altera-lo. Como Funciona isso?
Um bloco é composto por:
<ul>
    <li>Dados</li>
    <li>Hash</li>
    <li>Hash do Bloco Anterior</li>
</ul>

Os dados gravados dependerão do tipo de blockchain. O bloco do bitcoin guradam os dados de ums transação como comprador, vendedor e quantidade de moedas.
O bloco também possui uma hash que pode ser pensado com uma impressão digital que identifica o todo o seu conteudo e é único. Uma vez quem um bloco seja criado e sua hash calculada, mudar algo dentro do Bloco causará a mudança da hash, em outras palavras, hash são muito úteis para detectar mudanças nos blocos.

O terceiro elemento dentro do bloco é a hash do bloco anterior e é isso que cria a corrente e permite o BlockChain ser tão seguro
<b>#2:19 - 2:22</b>

Mas usar Hash não é suficiente para prevenir ataques com intenção de modificar os dados. Os computadores hoje são muito mais rapidos e podem calcular milhares de hash por segundo. Com isso é possivel alterar um bloco re reclacular todos os blocos parar tornar seu BlockChain válido na rede.

Para mitigar isso o BlockChain tem algo chamado proof of work (Prova de Trabalho). É um mecanismo que desacelera a criação de novos blocos. No caso do bitcoin leva-se 10 minutos para calcular o proof of work e adicionar um novo bloco a corrente. Esse mecanismo torna dificil alterar os blocos, porque se alterar um bloco, significa ter que minerar todos os novos blocos gastando 10 minutos em cada bloco. A segurança do BlockChain vem do uso criativo de criptografia Hash e o mecanismo de proof of work.

Outra forma do BlockChain assegurar is dadis é sendo distribuido. Ao invés de ausar uma unidade central para gerênciar a corrente, o BlockChain usa uma rede peer-to-peer e todos tem permissão para entrar.

Quando alquêm se junta a rede, o usuário recebe uma cópia da corrente inteira do BlockChain. O node pode usar isso para verificar que tudo está em orderm.

Na criação de um novo Bloco cada node verifica se cada bloco da corrente foi alterado e só após tudo verificado cada node adiciona o novo bloco criado ao seu próprio BlockChain.

Todos os Nodes desss rede criam consenso. Eles concordam sobre qual bloco é válido e qual não é. Blocos alterados serão rejeitados por outros nodes na rede.

Para alterar com sucesso um blco é necessario alterar todos os blocos na corrente, refazer todas as proof of work para cada Bloco e tomar controle de mais de 50% da reder peer-to-peer. Só assim o bloco alterado torna-se o aceito por todos. Isso é quase impossivel de ser executado!

### Adicionando Uma Transação ao Bloco

Precisamos de uma forma de adicionar transações ao Bloco. Nosso método nova_transacao() é responsável por isso:

```python
class Blockchain:
    def nova_transacao(self, dono_anterior, dono_atual, valor):
        """
        :param dono_anterior: <str> endereço do dono_anterior
        :param dono_atual: <str> endereço do novo_atual
        :param valor: <int> Valor
        :return: <int> O indice do Bloco que guardará essa transação
        """
        self.transacao_atual.append({
            'dono_anterior':dono_anterior,
            'dono_atual':dono_atual,
            'valor':valor,
         })
        return self.ultimo_bloco['indice'] + 1
```


## 6 - Hash e Como funciona o Proof of Work

Nessa altura podemos definir o BlockChain como uma corrente sequencial, imutável de registros chamados de blocos. Eles podem conter, alem de transações, arquivos ou qualquer dado que quiser. O importante é que ele sejam ligados em corrente usando Hashes

Funções hash são funções que recebem um valor de entrada e devolvem um resultado criptografado. Funções hash geralmente são funções de uma via, ou seja não é possível  descobrir o valor da entrada só sabendo o resultado:

Exemplo com função hash md5

```python
import hashlib
resposta = '42'
MD5 = hashlib.md5(resposta.encode()).hexdigest()
print(MD5)
```

O proof-of-work significa descobrir um número que resolve um problema, que pode ser um padrão no string resultante de uma função hash. O número tem que ser difícil de encontrar, mas fácil de verificar por um qualquer na rede.

Exemplo simples para ajudar.

Digamos que nosso problema a ser resolvido seja de que a função hash da multiplicação entre dois números x e y tem que terminar em 0. Então $hash(x . y)$ = abc23dc...0. Para ficar mais simples fixamos x = 5. Implementando no Python:

```python
x = 5
y = 0  # não sabemos os y ainda

while hashlib.sha256(f'{x*y}'.encode()).hexdigest()[-1] != "0":
    y += 1

print(y)
```

Com uma função hash sha256 criaremos uma hash para cada bloco e assim qualquer mudança de string dentro do bloco, produzirá uma diferente hash:

```python
class Blockchain:
    @staticmethod
        def hash(bloco):
            """
            Cria uma hash sha-256 para o bloco
            :param bloco: Bloco
            """

            # O dicionario tem que estar ordenado se não haverá incosistencia
            bloco_string = json.dumps(bloco, sort_keys=True).encode()
            return hashlib.sha256(bloco_string).hexdigest()
```

O algorítimo de proof-of-work do Bitcoin é chamado de Hashcash. Não é muito diferente do exemplo básico acima. É esse algorítimo que os mineradores disputam para resolver e de criar um novo bloco. Em geral, a dificuldade é determinada pelo número de caracteres procurados na string. Os mineradores são remunerados com uma parte da transação - coin. A rede é capaz de facilmente verificar a solução.

## 7 - Implementando o Proof of Work

Vamos colocar um algoritmo similar na nossa classe BlockChain e que terá o seguinte problema pra resolver:
<ul>Encontre um número que o resultado da função Hash aplicado sobre a junção da hash do bloco anterior com a atual tenha quatro  '0' zeros no lado direito da string.</ul>

```python
def proof_of_work(self, ultimo_bloco):
    """
    Simples algoritimo de prova de Trabalho:
    - Ache um número p' tal que hash(pp') contem 4 zeros do na ponta direita do resultado
    - Onde p é a prova anterior, e p' é a nova prova

    :param ultimo_bloco: <dict> ultimo bloco
    :return: <int>
    """
    ultima_prova = ultimo_bloco['prova']
    ultima_hash = self.hash(ultimo_bloco)

    prova = 0
    while self.prova_valida(ultima_prova, prova, ultima_hash) is False:
        prova += 1

    return prova

@staticmethod
def prova_valida(ultima_prova, prova, ultima_hash):
    """
    Validar a Prova de trabalho
    :param ultima_prova: <int> Prova Anterior
    :param prova: <int> Prova Atual
    :param ultima_hash: <str> A hash do bloco anterior
    :return: <bool> True se correto, False se not.
    """
    chute = f'{ultima_prova}{prova}{ultima_hash}'.encode()
    chute_hash = hashlib.sha256(chute).hexdigest()
    return chute_hash[:4] == "0000"
```

Para ajustar a dificuldade do algorítimo, podemos modificar o número de zeros, mas 4 é suficiente. Teste em com mais zeros e verá que o tempo pra achar o resultado começará a aumentar.

## 10 - Implementar o Consenso

Temos um BlockChain básico que aceita transações e permite minerar novo blocos, mas o objetivo do BlockChain é ser descentralizado. Se eles são descentralizados, como nos certificaremos de que todos possuem os mesmo registros de correntes? Isso é chamado de problema do consenso e precisaremos de consenso se quisermos mais de um node na nossa rede.

###  Registrando novos Nodes

Antes de ser possivel implementar o algoritimo do Consenso, precisaremos de um uma forma do node saber sobre os nodes vizinhos na rede. Cada node na rede deve manter or resgistro de outros nos na rede. Assim, precisaremos de mais endpoits:

<ul>
    <li>/nodes/registrar</li> #aceitar umas lista de novos nodes em forma de URLs
    <li>/nodes/resolver</li> # implementar Algoritimo de Consenso - verificar se o node tem a corrente correta
</ul>

Precisaremos também modificar o construtO
r do blockchain e criar um método para registrar nodes:

```python
class Blockchain(object):
    def __init__(self):
        self.transacao_atual = []
        self.corrente = []
        self.nodes = set()

        # Criar o Bloco genesis
        self.novo_bloco(hash_anterior='1', prova=100)
    
    def registrar_node(self, endereco):
        """
        Adiciona um novo endereço a lista de Nodes
        :param endereço: endereço do node. Ex. 'http://192.168.0.5:5000'
        """

        parsed_url = urlparse(endereco)
        if parsed_url.netloc:
            self.nodes.add(parsed_url.netloc)
        elif parsed_url.path:
            # Isso é para aceitar url alem de IP como '192.168.0.5:5000'.
            self.nodes.add(parsed_url.path)
        else:
            raise ValueError('URL invalida')
```

A função set() é utilizada para assegurar que não incluímos o mesmo endereço duas vezes. Não importa quantas vezes inserimos um específico node, aparecerá um resgistro. 

### Consenso

Um conflito é quando um node tem diferente correntes outros nodes. Para resolver isso, nós criaremos uma regra que a corrente mais longa da rede é a autoridade. Usando o algorítimo nós chemgamos ao consenso entre os nossos nodes da rede.

```python
import requests


class Blockchain(object)

    def corrente_valida(self, corrente):
        """
        Determina se um dado Bloco é válido
        :param corrente: um blockchain
        :return: True se valido, False se não
        """

        ultimo_bloco = corrente[0]
        indice_atual = 1

        while indice_atual < len(corrente):
            bloco = corrente[indice_atual]
            print(f'{ultimo_bloco}')
            print(f'{bloco}')
            print("\n-----------\n")
            # verifica de a hash do Bloco é correta
            hash_ultimo_bloco = self.hash(ultimo_bloco)
            if bloco['hash_anterior'] != hash_ultimo_bloco:
                return False

            # verifica se o proof_of_work é correto
            if not self.prova_valida(ultimo_bloco['prova'], bloco['prova'], hash_ultimo_bloco):
                return False

            ultimo_bloco = bloco
            indice_atual += 1

        return True

    def resolver_conflitos(self):
        """
        Esse é o nosso algoritimo de consenso que altera a nossa
        corrente com a maior da rede
        :return: True se for alterado, False se não
        """

        vizinhos = self.nodes
        nova_corrente = None

        # Estamos procurando uma corrente maior que a nossa na rede
        tamanho_max = len(self.corrente)

        # Pegar e verificar cada corrente da rede
        for node in vizinhos:
            resposta = requests.get(f'http://{node}/corrente')

            if resposta.status_code == 200:
                tamanho = resposta.json()['quantidade de ']
                corrente = resposta.json()['corrente']

                # Verifica se a corrente é a maior e se é valida
                if tamanho > tamanho_max and self.corrente_valida(corrente):
                    tamanho_max = tamanho
                    nova_corrente = corrente

        # Substituir nossa corrente se descobrirmos uma maior que a nossa
        if nova_corrente:
            self.corrente = nova_corrente
            return True

        return False
```

O primeiro método corrente_valida() executa um loop em cada bloco verificando cada prova e hash.

resolver_conflitos() é o método que acessará cada node da rede em busca da maior corrente da rende com o seguinte método:
Se a corrente encontrada é válida, e a corrente é maior que a nossa, troque a nossa.

Vamos adicionar os dois endpoints para a nossa API, uma para adicionar nodes vizinhos e outra para resolver conflitos:

```python
@app.route('/nodes/registrar', methods=['POST'])
def registrar_nodes():
    valores = request.get_json()

    nodes = valores.get('nodes')
    if nodes is None:
        return "Erro: Por Favor indique uma lista valida de nodes", 400

    for node in nodes:
        blockchain.registrar_node(node)

    resposta = {
        'mensagem': 'Novo node foi adicionado',
        'total_nodes': list(blockchain.nodes),
    }
    return jsonify(resposta), 201


@app.route('/nodes/resolver', methods=['GET'])
def consenso():
    substituidos = blockchain.resolver_conflitos()

    if substituidos:
        resposta = {
            'mensagem': 'Nossa Corrente foi Substituida',
            'nova_corrente': blockchain.corrente
        }
    else:
        response = {
            'mensagem': 'Nossa Corrente é a Mandataria',
            'Corrente': blockchain.corrente
        }

    return jsonify(resposta), 200
```
